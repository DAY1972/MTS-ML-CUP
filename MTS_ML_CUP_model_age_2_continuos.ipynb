```python
def convert_sparse_matrix_to_sparse_tensor(X):
  coo = X.tocoo()
  indices = np.mat([coo.row, coo.col]).transpose()
  return tf.SparseTensor(indices, coo.data, coo.shape)
```

```python
# define sets of inputs
input_count = tf.keras.Input(shape=(X_count_train.shape[1],), sparse=None)
input_city_name = tf.keras.Input(shape=(X_city_name_train.shape[1],), sparse=None)
input_cpe_manufacturer_name = tf.keras.Input(shape=(X_cpe_manufacturer_name_train.shape[1],), sparse=True)
input_cpe_model_name = tf.keras.Input(shape=(X_cpe_model_name_train.shape[1],), sparse=True)
input_cpe_model_os_type = tf.keras.Input(shape=(X_cpe_model_os_type_train.shape[1],), sparse=True)
input_cpe_type_cd = tf.keras.Input(shape=(X_cpe_type_cd_train.shape[1],), sparse=True)
input_region_name = tf.keras.Input(shape=(X_region_name_train.shape[1],), sparse=True)
input_text = tf.keras.Input(shape=(X_text_train.shape[1],), sparse=True)
input_url_host = tf.keras.Input(shape=(X_url_host_train.shape[1],), sparse=True)
input_url_host_all = tf.keras.Input(shape=(X_url_host_all_train.shape[1],), sparse=True)
```

```python
# branch count
x_count = tf.keras.layers.Dense(64, activation="relu")(input_count)
x_count = tf.keras.Model(inputs=input_count, outputs=x_count)
 
# branch city_name
x_city_name = tf.keras.layers.Dense(64, activation="relu")(input_city_name)
x_city_name = tf.keras.Model(inputs=input_city_name, outputs=x_city_name)

# branch cpe_manufacturer_name
x_cpe_manufacturer_name = tf.keras.layers.Dense(42, activation="relu")(input_cpe_manufacturer_name)
x_cpe_manufacturer_name = tf.keras.Model(inputs=input_cpe_manufacturer_name, outputs=x_cpe_manufacturer_name)

# branch cpe_model_name
x_cpe_model_name = tf.keras.layers.Dense(64, activation="relu")(input_cpe_model_name)
x_cpe_model_name = tf.keras.Model(inputs=input_cpe_model_name, outputs=x_cpe_model_name)

# branch cpe_model_os_type
x_cpe_model_os_type = tf.keras.layers.Dense(2, activation="relu")(input_cpe_model_os_type)
x_cpe_model_os_type = tf.keras.Model(inputs=input_cpe_model_os_type, outputs=x_cpe_model_os_type)

# branch cpe_type_cd
x_cpe_type_cd = tf.keras.layers.Dense(4, activation="relu")(input_cpe_type_cd)
x_cpe_type_cd = tf.keras.Model(inputs=input_cpe_type_cd, outputs=x_cpe_type_cd)

# branch region_name
x_region_name = tf.keras.layers.Dense(64, activation="relu")(input_region_name)
x_region_name = tf.keras.Model(inputs=input_region_name, outputs=x_region_name)

# branch text
x_text = tf.keras.layers.Dense(64, activation="relu",  kernel_regularizer='l2')(input_text)
x_text = tf.keras.Model(inputs=input_text, outputs=x_text)

# branch url_host
x_url_host = tf.keras.layers.Dense(64, activation="relu",  kernel_regularizer='l2')(input_url_host)
x_url_host = tf.keras.Model(inputs=input_url_host, outputs=x_url_host)

# branch url_host_all
x_url_host_all = tf.keras.layers.Dense(64, activation="relu",  kernel_regularizer='l2')(input_url_host_all)
x_url_host_all = tf.keras.Model(inputs=input_url_host_all, outputs=x_url_host_all)

# combine the output of the branches
combined = tf.keras.layers.concatenate([x_count.output,
                                        x_city_name.output,
                                        x_cpe_manufacturer_name.output,
                                        x_cpe_model_name.output,
                                        x_cpe_model_os_type.output,
                                        x_cpe_type_cd.output,
                                        x_region_name.output,
                                        x_text.output,
                                        x_url_host.output,
                                        x_url_host_all.output
                                        ])
 
# apply a FC layer and then a regression prediction on the combined outputs
z = tf.keras.layers.Dense(256, activation="relu")(combined)
z = tf.keras.layers.Dense(128, activation="relu")(z)
z = tf.keras.layers.Dense(64, activation="relu")(z)
z = tf.keras.layers.Dense(32, activation="relu")(z)
z = tf.keras.layers.Dense(16, activation="relu")(z)
z = tf.keras.layers.Dense(8, activation="relu")(z)
z = tf.keras.layers.Dense(4, activation="relu")(z)
z = tf.keras.layers.Dense(1, activation="linear")(z)
```

```python
# our model will accept the inputs of the branches and then output a single value
model = tf.keras.Model(inputs=[x_count.input,
                               x_city_name.input,
                               x_cpe_manufacturer_name.input,
                               x_cpe_model_name.input,
                               x_cpe_model_os_type.input,
                               x_cpe_type_cd.input,
                               x_region_name.input,
                               x_text.input,
                               x_url_host.input,
                               x_url_host_all.input
                               ], outputs=z)
```

4 эпох Adam + 6 эпох Adadelta learning_rate=0.1 + 10 эпох Adadelta learning_rate=0.01

```python
#opt = tf.keras.optimizers.Adam()
opt = tf.keras.optimizers.Adadelta(learning_rate=0.01)
LOSS = "mean_absolute_error"
metric = "mean_squared_error"
model.compile(optimizer=opt, loss=LOSS, metrics=metric)
callbacks = [tf.keras.callbacks.ModelCheckpoint(filepath="/content/drive/My Drive/MTS ML CUP/age_2_continuos.keras",
                                                save_best_only=True, monitor="val_loss")]
```

```python
history = model.fit([X_count_train,
                     X_city_name_train,
                     X_cpe_manufacturer_name_train,
                     X_cpe_model_name_train,
                     X_cpe_model_os_type_train,
                     X_cpe_type_cd_train,
                     X_region_name_train,
                     X_text_train,
                     X_url_host_train,
                     X_url_host_all_train
                     ],
                    y_train,
                    validation_data=([X_count_val,
                                      X_city_name_val,
                                      X_cpe_manufacturer_name_val,
                                      X_cpe_model_name_val,
                                      X_cpe_model_os_type_val,
                                      X_cpe_type_cd_val,
                                      X_region_name_val,
                                      X_text_val,
                                      X_url_host_val,
                                      X_url_host_all_val
                                      ],
                                     y_val),
                    verbose=1,
                    epochs=10,
                    batch_size=512,
                    callbacks=callbacks)
```
